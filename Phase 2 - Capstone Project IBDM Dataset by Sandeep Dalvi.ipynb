# SQLite Database

```python
import sqlite3
import pandas as pd
```

```python
def create_tables_and_insert_data():
    # Connecting to the SQLite database
    conn = sqlite3.connect('imdb_movies.db')
    cursor = conn.cursor()

    # Creating the first table for IMDbMoviesInfo.csv
    cursor.execute('''
        CREATE TABLE IF NOT EXISTS IMDbMoviesInfo (
            sno INTEGER PRIMARY KEY,
            "Movie Name" TEXT,
            Director1 TEXT,
            Director2 TEXT,
            Duration INTEGER,
            Year INTEGER,
            Ratings REAL,
            Metascore INTEGER)''')

    # Creating the second table for IMDbMoviesDetails.csv
    cursor.execute('''
        CREATE TABLE IF NOT EXISTS IMDbMoviesDetails (
            "Movie Name" TEXT,
            Votes INTEGER,
            "Gross Collection" TEXT,
            Popularity REAL,
            Certification TEXT,
            Star1 TEXT,
            Star2 TEXT,
            Star3 TEXT,
            Star4 TEXT,
            Genre1 TEXT,
            Genre2 TEXT,
            Genre3 TEXT)''')

    # Reading data from the CSV files
    csv1_df = pd.read_csv('IMDbMoviesInfo.csv')
    csv2_df = pd.read_csv('IMDbMoviesDetails.csv')
    
    # Insert data into the first table
    csv1_df.to_sql('IMDbMoviesInfo', conn, if_exists='replace', index=False)

    # Insert data into the second table
    csv2_df.to_sql('IMDbMoviesDetails', conn, if_exists='replace', index=False)
```
